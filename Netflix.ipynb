```python
import pandas as pd

df = pd.read_csv('/content/netflix.csv')

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
# Missing values fill cheyyadam

df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')

# Duplicate rows remove cheyyadam
df = df.drop_duplicates()

print("Missing Values after cleaning:")
print(df.isnull().sum())

print("\nShape after cleaning:")
print(df.shape)
```

```python
df['date_added'] = df['date_added'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')

print(df.isnull().sum())
```

```python
# Q1. Movies vs TV Shows
print(df['type'].value_counts())

# Q2. Top 10 Countries
print(df['country'].value_counts().head(10))

# Q3. Content Rating Distribution
print(df['rating'].value_counts())

# Q4. Top 10 Release Years
print(df['release_year'].value_counts().head(10))

# Q5. Top 10 Genres
print(df['listed_in'].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt

# Chart 1
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# Chart 2
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title('Top Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Chart 3
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Content Released by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()

# Chart 4
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()

# Chart 5
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Movies vs TV Shows Percentage')
plt.ylabel('')
plt.show()

# Chart 6
df['release_year'].plot(kind='hist', bins=20)
plt.title('Release Year Distribution')
plt.xlabel('Release Year')
plt.show()
```
